## Problem 1:

In general, if your data has $D$ columns, and you assume your model has $H$ clusters, how many parameters in total are there for

 - the general case of fully flexible covariance matrices,
 - assuming all groups have the same general covariance structure,
 - assuming variables are uncorrelated in all groups, and
 - assuming that in every group, all features have the same variance *and* are uncorrelated with one another?

Let D be the number of columns (features) and H be the number of clusters. The total number of parameters in each scenario is:

- **General case (fully flexible covariance matrices):**  
  Each group has a mean vector of length D and a D x D covariance matrix with D(D+1)/2 unique parameters.  
  Total: H × [D + D(D+1)/2] = H × (D + D(D+1)/2)

- **Same general covariance structure across all groups:**  
  One shared covariance matrix (D(D+1)/2 parameters), and H mean vectors.  
  Total: H × D + D(D+1)/2

- **Variables uncorrelated in all groups:**  
  Each covariance matrix is diagonal → D variances per group.  
  Total: H × (D + D) = H × 2D

- **Same variance and uncorrelated in all groups:**  
  One variance shared across all features and groups (1 parameter), and H mean vectors.  
  Total: H × D + 1

### Exploratory Data Analysis and Predictor Transformations

To assess the distribution of each predictor variable, we began by plotting histograms.

The histograms revealed strong right skew in the distributions of `glucose`, `insulin`, and `s2`. To reduce skewness and stabilize variance, we applied a log transformation:

- We used the transformation `log1p(x)` instead of `log(x)` to accommodate zero values and avoid `-inf` issues.
- This transformation is particularly helpful for Gaussian mixture modeling, where the assumption of roughly normal distributions within clusters is important.

The transformed variables were then used as input to the finite mixture model to improve cluster separation and model performance.

```python
# Apply log1p transformations to skewed predictors
diab["log_glufast"] = np.log1p(diab["glufast"])
diab["log_glutest"] = np.log1p(diab["glutest"])
diab["log_instest"] = np.log1p(diab["instest"])

# Select predictors and standardize them
X = diab[["relwt", "log_glufast", "log_glutest", "log_instest", "sspg"]]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title("PCA projection")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Run PCA on scaled data
pca = PCA(n_components=3)
X_pca_3d = pca.fit_transform(X_scaled)

# Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2], s=40, alpha=0.5)
ax.set_title("3D PCA Projection")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()
```

```python
bic_scores = []
aic_scores = []
k_range = range(1, 7)

for k in k_range:
    gmm = GaussianMixture(n_components=k, covariance_type='full').fit(X_scaled)
    bic_scores.append(gmm.bic(X_scaled))
    aic_scores.append(gmm.aic(X_scaled))

plt.plot(k_range, bic_scores, label='BIC')
plt.plot(k_range, aic_scores, label='AIC')
plt.xlabel("Number of Components")
plt.ylabel("Score")
plt.legend()
plt.title("Model Selection via AIC/BIC")
plt.show()
```

### Principal Component Analysis (PCA) and Cluster Selection

#### PCA Visualization of Transformed Data

To explore the underlying structure of the data, we performed Principal Component Analysis (PCA) on the standardized, transformed predictors. The PCA projection allowed us to visualize potential clusters in a lower-dimensional space.

**2D PCA Plot (First Two Principal Components)**  
The 2D plot shows two well-separated clusters along the first principal component, suggesting natural groupings in the data.

**3D PCA Plot (First Three Principal Components)**  
The 3D plot provides additional perspective, reinforcing the separation seen in 2D and indicating that most variation is captured within the first two or three components. This suggests that a relatively small number of dimensions is sufficient to characterize the group structure.

#### Model Selection using Information Criteria

We fit Gaussian Mixture Models (GMMs) across a range of cluster counts (K = 1 through 6) and evaluated model fit using **Bayesian Information Criterion (BIC)** and **Akaike Information Criterion (AIC)**.

**BIC/AIC Plot**  
Both BIC and AIC decreased sharply from K = 1 to K = 2, indicating improved model fit. However, after K = 2, the improvements plateaued, and BIC in particular began to rise, reflecting increased model complexity without substantial gain in explanatory power.

#### Cluster Count Selection

Based on these diagnostics, we selected **K = 2** clusters for the finite mixture model. This decision is supported by:
- Somewhat intuitive visual separation in the PCA plots.
- BIC minimization at K = 2, suggesting a good trade-off between fit and parsimony.
- Theoretical expectation of two primary metabolic states in this data (e.g., insulin resistant vs. non-resistant), aligning with domain knowledge.

This cluster count provided a balance between interpretability, statistical justification, and visual coherence in the data.

```python
# Create DataFrame with proper column names
X_df = pd.DataFrame(X_scaled, columns=[f"x{i+1}" for i in range(X_scaled.shape[1])])

# Fit KMeans using all 5 features
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Add cluster labels to the DataFrame
X_df["label"] = labels

# View cluster-wise summary statistics
summary = X_df.groupby("label").agg(["mean", "std"])
print(summary)
```

```python
# Reduce to 2D for plotting
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a scatter plot with label-based coloring
plt.figure(figsize=(8, 6))
for lbl in [0, 1]:
    plt.scatter(
        X_pca[labels == lbl, 0], 
        X_pca[labels == lbl, 1], 
        label=f"Cluster {lbl}", 
        alpha=0.7
    )

plt.title("KMeans Clustering Results (PCA-reduced data)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.tight_layout()
plt.show()
```

To address the label-switching issue commonly observed in traceplots for finite mixture models with exchangeable priors, we used KMeans clustering as a preprocessing step to guide prior specification. By clustering the data into two groups and examining the summary statistics (means and standard deviations) of each cluster, we were able to identify distinguishing features between the groups.

Using this insight, we set **informative and asymmetric priors** on the component means (`mu`) and standard deviations (`sigma`) in the Stan model. For example, one cluster was centered tightly around lower values, while the other was more spread out and centered around higher values for certain variables. Encoding this structure into the priors (e.g., `mu[1] ~ normal(-2, 1)` and `mu[2] ~ normal(2, 1)`) helped anchor the clusters and **prevented the sampler from switching group labels between chains**, thus resolving the label-switching issue and improving the interpretability of the posterior output.

```python
# Prepare data for Stan
stan_data = {
    "N": X_scaled.shape[0],   # number of observations
    "D": X_scaled.shape[1],   # number of predictors
    "K": 2,                   # number of mixture components
    "x": X_scaled             # predictor matrix
}

# Compile and fit the model
model = CmdStanModel(stan_file="finite_mixture_normals.stan")
fit = model.sample(
    data=stan_data
)
```

```python
print(fit.diagnose())
```

```python
# Parameters to plot
params = ['lambda[1]', 'lambda[2]', 'mu[1,1]', 'mu[2,1]', 'sigma[1,1]', 'sigma[2,1]']
chains = fit.chains
posterior = fit.draws_pd()

# Trace plots
fig, axs = plt.subplots(len(params), 1, figsize=(10, 2.5 * len(params)), sharex=True)

for i, param in enumerate(params):
    for chain_id in range(chains):
        chain_samples = posterior[param].values[chain_id::chains]
        axs[i].plot(chain_samples, alpha=0.3, color='blue')  # fixed color for all chains
    axs[i].set_ylabel(param)
    axs[i].set_title(f"Trace Plot for {param}")

plt.xlabel("Iteration")
plt.tight_layout()
plt.show()
```

### Trace Plots and Parameter Selection

The above trace plots show the sampled values over iterations for selected model parameters: the cluster weights (`lambda[1]`, `lambda[2]`), one dimension of the cluster means (`mu[1,1]`, `mu[2,1]`), and one dimension of the cluster standard deviations (`sigma[1,1]`, `sigma[2,1]`). We chose these representative parameters to examine convergence behavior across both mixture proportions and component-specific location and scale parameters.

In our case, the fuzzy, overlapping patterns across all chains — with no visible stickiness, drift, or abrupt shifts — indicate strong mixing and stable estimates. This visual inspection supports the conclusions from the formal diagnostics (e.g., R-hat near 1.0, no divergent transitions), increasing our confidence in the validity of the posterior samples for downstream inference.

**2. What did you spend most of your time doing to get this to be so?**

### Effort Spent on Ensuring Identifiability

Most of the time was spent diagnosing and resolving **label switching** and **non-identifiability** issues common to finite mixture models with exchangeable priors.

Key steps included:

- **Diagnosing poor mixing** and multimodal trace plots in early model runs.
- **Exploring cluster structure** with KMeans and PCA to understand the natural separation in the data.
- **Designing informative, asymmetric priors** for the component parameters ($\mu_k$, $\sigma_k$) based on KMeans cluster summaries, breaking exchangeability and anchoring each mixture component to a specific mode in the posterior.
- **Rewriting the Stan model** to incorporate these asymmetric priors and stabilize the sampling behavior.
- **Verifying convergence diagnostics** (e.g., $\hat{R}$, ESS, trace plots) after each change to confirm improved posterior geometry and identifiability.

This iterative process was essential to overcome the symmetric nature of exchangeable priors and enforce a unique labeling of mixture components.

**4. Should you pay attention to $\hat{R}$ diagnostics for `label_prob` variables?**

No. The label_prob variables are generated quantities derived from the posterior samples of the model parameters ($\mu$, $\sigma$, $\lambda$). As such, they are not directly sampled in the MCMC process and do not affect convergence of the Markov chains.

$\hat{R}$ statistics are only meaningful for parameters that are explicitly sampled in the parameters block. Since label_prob is computed post hoc from already-sampled parameters, any $\hat{R}$ values reported for it should be ignored.

**5. Which individual is one of the most difficult to categorize? Why do you think this is interesting to identify?**

```python
# Extract posterior samples for label probabilities (shape: draws x N x K)
label_prob_samples = fit.stan_variable("label_prob")  # shape: (draws, N, K)

# Compute posterior mean label probabilities for each observation
label_prob_mean = label_prob_samples.mean(axis=0)  # shape: (N, K)

# Compute uncertainty as distance from 0.5
uncertainty = np.abs(label_prob_mean[:, 0] - 0.5)

# Identify the index of the most uncertain observation
most_uncertain_idx = np.argmin(uncertainty)
most_uncertain_probs = label_prob_mean[most_uncertain_idx]

print(f"Most uncertain individual is at index {most_uncertain_idx}")
print(f"Posterior label probabilities: Cluster 0 = {most_uncertain_probs[0]:.3f}, Cluster 1 = {most_uncertain_probs[1]:.3f}")

# Inspect their feature values
X_df.iloc[most_uncertain_idx]
```

Individual 114 was the most difficult to categorize, with posterior label probabilities of 0.414 for Cluster 0 and 0.586 for Cluster 1—indicating substantial uncertainty. This is interesting because it highlights observations that do not cleanly belong to a single latent group and may share characteristics of both. Identifying such individuals can inform potential overlap between groups, the need for more flexible modeling, or real-world heterogeneity that merits further investigation.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot PCA with clusters
plt.figure(figsize=(8, 6))
for lbl in [0, 1]:
    plt.scatter(
        X_pca[labels == lbl, 0], 
        X_pca[labels == lbl, 1], 
        label=f"Cluster {lbl}", 
        alpha=0.7
    )

# Highlight the most uncertain individual
plt.scatter(
    X_pca[most_uncertain_idx, 0], 
    X_pca[most_uncertain_idx, 1], 
    color='black', 
    edgecolor='white', 
    marker='*', 
    s=200, 
    label='Most Uncertain'
)

plt.title("PCA of Input Data with Most Uncertain Individual Highlighted")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Principal Component Plot of Groups vs. Inferred Labels

The following confusion matrix compares the true labels (rows) with the inferred cluster labels (columns):

This comparison helps validate the model's practical utility while also highlighting limitations when mapping unsupervised clusters to clinical categories.

```python
# Step 1: Map original group to binary labels (0 = Normal, 1 = Diabetic)
diab_binary = diab["group"].map(lambda g: 0 if g == "Normal" else 1).values

# Step 2: Inferred labels from clustering (already in X_df["label"])
# Try both label alignments to find the best match
match_direct = np.mean(diab_binary == X_df["label"].values)
match_flipped = np.mean(diab_binary == (1 - X_df["label"].values))

# Choose best alignment
best_match = max(match_direct, match_flipped)
alignment = "direct" if match_direct >= match_flipped else "flipped"

# Print result
print(f"Best alignment: {alignment}")
print(f"Percentage agreement with original group labels: {best_match * 100:.2f}%")
```

```python
# Map true labels to binary: 0 = Normal, 1 = Diabetic
true_labels = diab["group"].map(lambda g: 0 if g == "Normal" else 1).values
inferred_labels = X_df["label"].values

# Set up the figure
plt.figure(figsize=(8, 6))

# Define colors and markers manually
colors = ["red", "blue"]  # 0 = Normal, 1 = Diabetic
markers = ["o", "X"]      # 0 = Cluster 0, 1 = Cluster 1

# Plot each combination
for true in [0, 1]:
    for inferred in [0, 1]:
        mask = (true_labels == true) & (inferred_labels == inferred)
        plt.scatter(
            X_pca[mask, 0],
            X_pca[mask, 1],
            c=colors[true],
            marker=markers[inferred],
            edgecolor='k',
            label=f"True: {true}, Inferred: {inferred}",
            alpha=0.7,
            s=70
        )

# Legend and labels
plt.title("True vs Inferred Labels (Color = True, Marker = Inferred)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Label Mapping")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**7. Are there any interesting scientific questions that are not addressed by this finite mixture model?**

Yes. While the finite mixture model effectively uncovers latent structure in the data, it does not account for temporal dynamics, causal relationships, or covariate interactions. For example, it does not explain *why* individuals fall into different metabolic states or how variables like insulin resistance and glucose levels evolve over time. Additionally, the model assumes a fixed number of clusters and does not consider uncertainty in the number of subpopulations, which could be addressed using nonparametric Bayesian methods. Finally, it treats all variables symmetrically without modeling domain knowledge (e.g., physiological mechanisms), limiting interpretability and causal inference.

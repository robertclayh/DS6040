## Problem 2:

Fit the following model. Discuss the posterior you found in detail. 

 - $y$ is the monthly sum of `drivers`, `front`, and `rear`.
 - $x_1$ is `PetrolPrice`
 - $x_2$ is the `law` dummy variable

$$
\log(\lambda) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 
$$

$$
y \mid x_1, x_2 \sim \text{Poisson}(\lambda)
$$

```python
# Create response variable
road_data["y"] = road_data["drivers"] + road_data["front"] + road_data["rear"]

# Predictor variables
x1 = road_data["PetrolPrice"].values
x2 = road_data["law"].values
y = road_data["y"].values

# Design matrix with intercept manually handled in Stan
X = np.column_stack([x1, x2])

# Stan data
stan_data = {
    "N": len(y),
    "K": X.shape[1],
    "X": X,
    "y": y
}
```

```python
model = CmdStanModel(stan_file='poisson_regression.stan')

fit = model.sample(
    data=stan_data,
    chains=4,
    parallel_chains=4,
    iter_warmup=500,
    iter_sampling=1000,
    seed=42
)
```

```python
posterior = fit.draws_pd()

# Summarize key parameters
posterior[["alpha", "beta[1]", "beta[2]"]].describe(percentiles=[0.05, 0.5, 0.95])
```

### Posterior Summary: Poisson Regression Model

We fit a Poisson regression model to predict the monthly number of car driver and passenger casualties using two predictors: `PetrolPrice` and a `law` indicator for seatbelt legislation.

- **Intercept (alpha)**: The posterior mean of 8.50 represents the expected log count of casualties when both `PetrolPrice` and `law` are zero. This is not directly interpretable but serves as a baseline.

- **PetrolPrice (beta[1])**: The coefficient has a posterior mean of -4.87 with a 90% credible interval of [-5.07, -4.67]. Since the model is on the log scale, this corresponds to a strong inverse relationship: higher petrol prices are associated with significantly fewer casualties. Specifically, a one-unit increase in `PetrolPrice` multiplies the expected count by `exp(-4.87) ≈ 0.0076`, indicating a steep drop.

- **Seatbelt Law (beta[2])**: The posterior mean is -0.19, with a tight 90% credible interval of [-0.198, -0.182]. This implies that the implementation of seatbelt laws is associated with a roughly **17% decrease** in expected monthly casualties, since `exp(-0.19) ≈ 0.83`.

All posterior distributions are unimodal and show tight credible intervals, indicating strong signal and good model convergence.


```python
posterior[["alpha", "beta[1]", "beta[2]"]].hist(bins=30, figsize=(12, 4), layout=(1, 3), edgecolor='black')
plt.suptitle('Posterior Distributions of Poisson Model Parameters')
plt.tight_layout()
plt.show()
```

The posterior histograms for `alpha`, `beta[1]`, and `beta[2]` confirm that the MCMC chains mixed well and that the parameter distributions are unimodal and approximately symmetric. All three distributions are tightly concentrated, indicating high certainty in the estimates. The distribution for `beta[2]`, which corresponds to the seatbelt law, is clearly negative and well-separated from zero, supporting the conclusion that the law had a meaningful effect in reducing casualties.

## Problem 2:

Come up with some posterior (predictive) summary that suggests you might be missing a predictor that has to do with time!

```python
# Extract posterior draws
alpha_draws = posterior['alpha'].values
beta1_draws = posterior['beta[1]'].values
beta2_draws = posterior['beta[2]'].values

# Predictors
X = stan_data['X']
N = X.shape[0]

# Draw samples
n_draws = len(alpha_draws)
y_pred = np.zeros((n_draws, N))

for i in range(n_draws):
    lambda_i = np.exp(alpha_draws[i] + X @ np.array([beta1_draws[i], beta2_draws[i]]))
    y_pred[i, :] = np.random.poisson(lam=lambda_i)
```

```python
y_mean_pred = y_pred.mean(axis=0)  # posterior predictive mean for each month
y_actual = stan_data["y"]

plt.figure(figsize=(14, 5))
plt.plot(y_actual, label='Actual y', marker='o')
plt.plot(y_mean_pred, label='Posterior Predictive Mean', marker='x')
plt.title('Actual vs Posterior Predictive Means Over Time')
plt.xlabel('Time (Months)')
plt.ylabel('Casualties')
plt.legend()
plt.tight_layout()
plt.show()
```

## Problem 4:

Another way to capture a seasonal affect is to use a sine/cosine wave. This does not require a separate predictor for every month. Consider the following function:

$$
A \cos(2 \pi t \frac{1}{12} + \phi)
$$

 - $A$ is the *amplitude* that stretches to standard cosine wave up or down
 - $1/12$ is the *frequency*. In this case, a cycle is made every $12$ observations (we have monthly data).
 - $\phi$ is the *phase* that adjusts the wave to start at something different than $1$
 - $t = 1,2,\ldots$ is time or in our case the row number

We can visualize this function as a function of $t$. Notice that the distance from peak to peak, from trough to trough is exactly the reciprocal of the frequency. The frequency flipped is called the *period*.

```python
A = 6
phi = 1/2
t = np.arange(1,48)
freq = 1/12
wave = A*np.cos(2*np.pi*freq*t + phi)
plt.plot(t,wave)
```

The unknown quantities are the phase and the amplitude. If we wanted to estimate these numbers using linear regression, we wouldn't be able to use this expression indirectly. The unknown coefficients are not betas and are not multiplied by the predictor in a linear way.

We could use a nonlinear regression, or we could cleverly use some clever trigonometry to rewrite the above function as

$$
\beta_{3} \cos(2 \pi t [1/12]) + \beta_{4}\sin(2 \pi t [1/12]).
$$

We'll add two predictors. One for $\cos(2 \pi t [1/12])$ and the other for $\sin(2 \pi t [1/12])$. 


Now I can ask you to fit the following model, and discuss the posterior you found. 

$$
\log(\lambda) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_{3} x_{3} + \beta_4 x_4
$$

$$
y \mid x_1, \ldots, x_{4} \sim \text{Poisson}(\lambda)
$$

 - $y$ is the monthly sum of `drivers`, `front`, and `rear`.
 - $x_1$ is `PetrolPrice`
 - $x_2$ is the `law` dummy variable
 - $x_3$ is $\cos(2 \pi t [1/12])$ with $t$ indicating the row number.
 - $x_{4}$ is $\sin(2 \pi t [1/12])$

```python
# Time index (1 to number of rows)
t = np.arange(1, len(road_data) + 1)
freq = 1/12

# Periodic predictors
road_data['cos_month'] = np.cos(2 * np.pi * freq * t)
road_data['sin_month'] = np.sin(2 * np.pi * freq * t)

# Prepare design matrix
X = road_data[['PetrolPrice', 'law', 'cos_month', 'sin_month']]

# Stan data
stan_data = {
    'N': len(road_data),
    'K': X.shape[1],
    'X': X.values,
    'y': road_data['y'].values.astype(int)
}
```

```python
model = CmdStanModel(stan_file='poisson_regression.stan')
fit = model.sample(data=stan_data, chains=4, parallel_chains=4, seed=42)
```

```python
# Load posterior draws
df = fit.draws_pd()

# Filter only parameters of interest (alpha and beta)
param_cols = ['alpha'] + [col for col in df.columns if col.startswith('beta')]
summary = df[param_cols].describe(percentiles=[0.05, 0.5, 0.95]).T

# Rename columns for clarity
summary = summary[['mean', 'std', '5%', '50%', '95%']]
summary.columns = ['Mean', 'SD', '5%', 'Median', '95%']

# Display summary table
summary
```

The fitted Poisson regression model incorporating a seasonal component via sine and cosine predictors produced well-behaved posterior summaries. The intercept term $ \alpha $ has a mean estimate of 8.487 with a narrow 90% credible interval ranging from 8.467 to 8.507, indicating high certainty about the baseline log-rate of serious injuries and deaths.

The coefficient on `PetrolPrice`, $ \beta_1 $, is estimated at -4.788 (90% CI: -4.983 to -4.591), suggesting that higher petrol prices are associated with fewer casualties, likely due to reduced driving. The coefficient on the `law` variable, $ \beta_2 $, is -0.191 (90% CI: -0.199 to -0.183), indicating a substantial decrease in the expected number of monthly injuries and fatalities following the introduction of seatbelt legislation.

Importantly, the seasonal wave terms also show strong effects. The cosine term $ \beta_3 $ has a positive coefficient of 0.051 (90% CI: 0.048 to 0.054), while the sine term $ \beta_4 $ is negative at -0.099 (90% CI: -0.102 to -0.096). These coefficients capture the seasonal fluctuation in casualty counts across the year, demonstrating that monthly variation can be effectively modeled using trigonometric predictors rather than separate dummy variables for each month. Overall, this model provides a compact and interpretable approach to incorporating seasonality into the accident rate estimation.

```python
# Plot settings
params_to_plot = ['alpha', 'beta[1]', 'beta[2]', 'beta[3]', 'beta[4]']
fig, axes = plt.subplots(len(params_to_plot), 2, figsize=(12, 3 * len(params_to_plot)))

for i, param in enumerate(params_to_plot):
    # Trace plot
    axes[i, 0].plot(df[param])
    axes[i, 0].set_title(f"Trace plot: {param}")
    axes[i, 0].set_xlabel("Iteration")
    axes[i, 0].set_ylabel("Value")
    
    # Density plot
    sns.kdeplot(df[param], ax=axes[i, 1], fill=True)
    axes[i, 1].set_title(f"Density plot: {param}")
    axes[i, 1].set_xlabel("Value")
    axes[i, 1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

### Trace and Density Plots: Convergence Diagnostics

The trace and density plots for each parameter ($\alpha$, $\beta_1$, $\beta_2$, $\beta_3$, and $\beta_4$) confirm good mixing and convergence of the MCMC chains.

## Problem 5:

We should also discuss **offsets** which is another trick you can use in Poisson regression. 

In Poisson regression, we estimate the varying $\lambda$. This is the average number of occurences of a count. Each row has a different average depending on what that row's predictors are.

However, sometimes one is interested or compelled to estimate a *rate*. 

A rate can be thought of as an average count **per unit of something else.** It can be average count *per unit time* or average count *per unit space* or something else. Then there is a simple transformation you can make to your model that will allow a Poisson regression to be employed. 

Let's suppose we are interest in estimating $\lambda / \text{time}$. Suppose $\text{time}$ is different for each row, and that's why we want to take it into account. 


If 
$$
\log\left(\frac{\lambda}{\text{time}}\right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 
$$

then

$$
\log\left(\lambda\right) = \log(\text{time}) + \beta_0 + \beta_1 x_1 + \beta_2 x_2 
$$

So, an **offset is a predictor that has a forced coefficient of $1$**.

NB: We aren't going to use time in our problem below--because every row has the same amount of time: one month--but this is just an assumption for my explanation.  


Fit the following model to estimate the average rate of accidents **per kilometer driven**. Discuss the posterior you found in detail. $$
\log\left(\frac{\lambda}{\text{kms}}\right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 
$$


 - $y$ is the monthly sum of `drivers`, `front`, and `rear`.
 - $x_1$ is `PetrolPrice`
 - $x_2$ is the `law` dummy variable
 - $\text{kms}$ is an offset relating to the `kms` column


```python
# Define response
road_data['y'] = road_data[['drivers', 'front', 'rear']].sum(axis=1)

# Define predictors
X = road_data[['PetrolPrice', 'law']].values

# Compute the offset as the log of kilometers driven
offset = np.log(road_data['kms'].values)

# Prepare data for Stan
stan_data = {
    "N": len(road_data),
    "K": X.shape[1],
    "X": X,
    "y": road_data["y"].values.astype(int),
    "log_kms": np.log(road_data["kms"].values)
}
```

### Problem 5: Poisson Regression with Offset (Rate per Kilometer)

In this model, we incorporated an **offset** to estimate the rate of serious accidents and fatalities **per kilometer driven**. This allows us to model $\log(\lambda / \text{kms})$ rather than raw counts, adjusting for exposure. Specifically, we model:

$$
\log(\lambda) = \log(\text{kms}) + \beta_0 + \beta_1 x_1 + \beta_2 x_2
$$

where:

- $y$ is the monthly sum of `drivers`, `front`, and `rear`.
- $x_1$ is `PetrolPrice`.
- $x_2$ is the `law` dummy variable.
- $\log(\text{kms})$ is the offset term.

#### Interpretation

### Proposed Final Model

We propose a final Poisson regression model that estimates the monthly number of serious injuries and fatalities in road accidents, adjusting for both exposure and seasonality. Specifically, we model the log of the expected count per kilometer driven as a function of fuel price, the implementation of a seatbelt law, and a seasonal cycle captured through a sinusoidal transformation. The model takes the form:

$$
\log(\lambda) = \log(\text{kms}) + \beta_0 + \beta_1 \cdot \text{PetrolPrice} + \beta_2 \cdot \text{Law} + \beta_3 \cdot \cos(2\pi t/12) + \beta_4 \cdot \sin(2\pi t/12)
$$

where $\lambda$ is the expected monthly count of accidents, and $t$ is the row index (representing month over time). Including $\log(\text{kms})$ as an offset allows us to model the rate of incidents per kilometer driven, making this model suitable for rate-based comparisons across months with differing driving exposure.

```python
# Add time index
road_data['t'] = np.arange(1, len(road_data) + 1)
freq = 1/12

# Create cosine and sine predictors
road_data['cos_12'] = np.cos(2 * np.pi * freq * road_data['t'])
road_data['sin_12'] = np.sin(2 * np.pi * freq * road_data['t'])

# Build predictor matrix
X = road_data[['PetrolPrice', 'law', 'cos_12', 'sin_12']]

# Stan data dictionary
stan_data = {
    "N": len(road_data),
    "K": X.shape[1],
    "X": X.values,
    "y": road_data[["drivers", "front", "rear"]].sum(axis=1).astype(int).values,
    "log_kms": np.log(road_data["kms"].values)
}

# Fit the model
model_final = CmdStanModel(stan_file="poisson_offset.stan")
fit_final = model_final.sample(data=stan_data, chains=4, seed=42)
```

```python
# Extract posterior draws from the fitted model
draws_df = fit_final.draws_pd()

# Select relevant parameter columns
summary_cols = ['alpha', 'beta[1]', 'beta[2]', 'beta[3]', 'beta[4]']

# Compute summary statistics: mean, SD, 5%, median, and 95%
summary_df = draws_df[summary_cols].describe(percentiles=[0.05, 0.5, 0.95]).T
summary_df = summary_df[['mean', 'std', '5%', '50%', '95%']]
summary_df.columns = ['Mean', 'SD', '5%', 'Median', '95%']

# Display the summary table
summary_df
```

```python
# Extract posterior draws
posterior = fit_final.draws_pd()[['alpha', 'beta[1]', 'beta[2]', 'beta[3]', 'beta[4]']]

# Set up the plot grid
n_params = posterior.shape[1]
fig, axes = plt.subplots(n_params, 1, figsize=(8, 2.5 * n_params), constrained_layout=True)

# Plot each parameter's posterior in its own subplot
for i, param in enumerate(posterior.columns):
    sns.kdeplot(posterior[param], ax=axes[i], fill=True, alpha=0.3)
    axes[i].set_title(f"Posterior Density: {param}")
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Density")

plt.show()
```

Our final model incorporates both exposure adjustment and seasonal variation by including a $\log(\text{kms})$ offset and sinusoidal terms for cyclical monthly effects. The **law** dummy variable ($\beta_2 \approx -0.41$) also shows a significant negative effect, supporting the conclusion that the seatbelt law reduced accident severity.
